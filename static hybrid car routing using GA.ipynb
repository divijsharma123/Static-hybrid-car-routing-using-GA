```python
# constant paramters for fitness function for cost , time, petrol and electricity repectively
import numpy as np
constants=[1,2,3,3]
parameters=np.array(constants)
```

```python
import random
# size of population to be kept each generation

pop_size=8
# size of chromosome
chromosome_size=4
# no. of generation till we want process to continue
no_of_generations=10
# no. of parents to be kept to produce off_springs 
no_of_parents=4
# generating random population of size 
r=[]
for i in range(pop_size*4):
    r.append(random.randint(0,1))

fuel_pop=np.array(r)
fuel_pop=fuel_pop.reshape((pop_size,4))
print(fuel_pop)
print()
r=[]
for i in range(pop_size*4):
    r.append(random.randint(1,3))

path_pop=np.array(r)
path_pop=path_pop.reshape((pop_size,4))

print(path_pop)





```

```python
def fitness_func(parameters,fuel_pop,path_pop,path1,path2,path3,path4):
    fitness=[]
    
    for i in range(fuel_pop.shape[0]):
        cost=path1[fuel_pop[i][0]][path_pop[i][0]]["cost"]+  path2[fuel_pop[i][1]][path_pop[i][1]]["cost"]  +path3[fuel_pop[i][2]][path_pop[i][2]]["cost"]  +path4[fuel_pop[i][3]][path_pop[i][3]]["cost"] 
        time=path1[fuel_pop[i][0]][path_pop[i][0]]["time"]+  path2[fuel_pop[i][1]][path_pop[i][1]]["time"]  +path3[fuel_pop[i][2]][path_pop[i][2]]["time"]  +path4[fuel_pop[i][3]][path_pop[i][3]]["time"] 
        petrol=path1[fuel_pop[i][0]][path_pop[i][0]]["petrol"]+  path2[fuel_pop[i][1]][path_pop[i][1]]["petrol"]  +path3[fuel_pop[i][2]][path_pop[i][2]]["petrol"]  +path4[fuel_pop[i][3]][path_pop[i][3]]["petrol"]
        electricity=path1[fuel_pop[i][0]][path_pop[i][0]]["electricity"]+  path2[fuel_pop[i][1]][path_pop[i][1]]["electricity"]  +path3[fuel_pop[i][2]][path_pop[i][2]]["electricity"]  +path4[fuel_pop[i][3]][path_pop[i][3]]["electricity"]
        variables=np.array([cost,time,petrol,electricity])
        fitness_i=sum(parameters*variables)
        fitness.append(fitness_i)
    
    fitness=np.array(fitness)
    return fitness
```

```python
def select_mating_pool(fitness,fuel_pop,path_pop,no_of_parents):

    # Selecting the best individuals in the current generation as parents for producing the offspring of the next generation.

    parents_fuel_pop= np.empty((no_of_parents, fuel_pop.shape[1]))
    parents_path_pop= np.empty((no_of_parents, fuel_pop.shape[1]))

    for parent_num in range(no_of_parents):

        min_fitness_idx = np.where(fitness == np.min(fitness))

        min_fitness_idx = min_fitness_idx[0][0]

        parents_fuel_pop[parent_num, :] = fuel_pop[min_fitness_idx, :]
        
        parents_path_pop[parent_num, :] = path_pop[min_fitness_idx, :]
    

        fitness[min_fitness_idx] =100000000

    return parents_fuel_pop,parents_path_pop
```

```python

def crossover(parents_fuel_pop,parents_path_pop, offspring_size):
    offspring_fuel_pop = np.empty(offspring_size)
    offspring_path_pop = np.empty(offspring_size)
    
     # The point at which crossover takes place between two parents. Usually, it is at the center.
    crossover_point = np.uint8(offspring_size[1]/2)
    
    for k in range(offspring_size[0]):
        # Index of the first parent to mate.
        parent1_idx = k%parents_fuel_pop.shape[0]
        
        
        # Index of the second parent to mate.
        parent2_idx = (k+1)%parents_fuel_pop.shape[0]
        
        # The new offspring will have its first half of its genes taken from the first parent.
        offspring_fuel_pop[k, 0:crossover_point] = parents_fuel_pop[parent1_idx, 0:crossover_point]
        offspring_path_pop[k, 0:crossover_point] = parents_path_pop[parent1_idx, 0:crossover_point]
        
        
        # The new offspring will have its second half of its genes taken from the second parent.
        offspring_fuel_pop[k, crossover_point:] = parents_fuel_pop[parent2_idx, crossover_point:]
        offspring_path_pop[k, crossover_point:] = parents_path_pop[parent2_idx, crossover_point:]
    
    return offspring_fuel_pop,offspring_path_pop
    
    
        
        
        
        
        
         
        
        
         
        
        
        
         
         
         
         
        
        
         
         
        
     
    
    
   
     
        
    
    
    
     
     
      
     
    
 
     
```

```python
def mutation(offspring_crossover_fuel_pop,offspring_crossover_path_pop):
    

    # Mutation changes a single gene in each offspring randomly.

    for idx in range(offspring_crossover_fuel_pop.shape[0]):
          # 2nd gene -: feul_mode is flipped and path is changed to random between 1 to 3
        if(offspring_crossover_fuel_pop[idx,2]==0):
            offspring_crossover_fuel_pop[idx,2]=1
        else:
            offspring_crossover_fuel_pop[idx,2]=0
        random_route=random.randint(1,3)
        offspring_crossover_path_pop[idx,2]=random_route
    
    
    return offspring_crossover_fuel_pop,offspring_crossover_path_pop
        
        
            
            

      
        

        
 
        

   
```
